# STUDENT MARK PREDICTOR

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = r"https://drive.google.com/uc?id=1xv853hZLOz82BNje5_g0JqWqpH7bxzQA&export=download"
df = pd.read_csv(path)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## Visualization the dataset to gain insight

```python
plt.scatter(x = df.study_hours, y = df.student_marks)
plt.xlabel("Student study hours")
plt.ylabel("Student marks")
plt.title("Scatter plot")
plt.show()
```

## preparing data for Machine learning model algorithms

```python
df.isnull()
```

```python
df.isnull().sum()
```

### so there are 5 data in study_hours which has null values

## now filling the null values with the mean value of the attribute

```python
df2 = df.fillna(df.mean())
```

```python
df2.isnull().sum()
```

```python
df2.head()
```

## Split the dataset into independent and dependent variable

```python
X = df2.drop("student_marks", axis ="columns")
y = df2.drop("study_hours", axis = "columns")
```

```python
print("shape of X",X.shape)
print("shape of y", y.shape)
```

## split the dataset into training and testing set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state =51)
```

```python
print("shape of X_train : ", X_train.shape)
print("shape of X_test : ", X_test.shape)
```

# Model Selection

## IN The visuliazation part we have seen that data are plotting in a linear manner , so here gonna use LinearRegression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression() 
# lr is the object of LinearRegresion
```

## fit the model

```python
lr.fit(X_train, y_train)
```

### so model is trained

```python
# A linear regression model works on y = mx+c;
#To find the value of m
print("value of m", lr.coef_)

#to find the c value
print("value of c", lr.intercept_)
```

## To test the model

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
# Measuring the Mean Absolute Error 

from sklearn import metrics
MAE = metrics.mean_absolute_error(y_test, y_pred)
MAE 
```

```python
# Model accuracy 

accuracy = lr.score(X_test, y_test)
accuracy*100
```

### So the model predict with more than 95% accuracy

```python
# to see the difference values

pd.DataFrame(np.c_[X_test, y_test, y_pred], columns= ["study_hours", "original_mark", "predicted_mark"])
```

```python
plt.scatter(X_test, y_test)
```

```python
plt.scatter(X_test, y_test)

plt.plot(X_train, lr.predict(X_train), color = 'r')
plt.grid()
```
